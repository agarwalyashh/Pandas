### Group By

Groups different categories and helps in analyzing data based on groups

```python
movies=pd.read_csv('movies.csv')
movies
```

```python
genres=movies.groupby('Genre')
genres.min()
```

```python
# Finding top 3 genres by total gross value
gross=movies.groupby('Genre')
gross['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
#Find genre with highest avg imdb rating
imdb=movies.groupby('Genre')
imdb['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# Find number of movies done by each actor
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)

#Alternatively -- movies['Star1'].value_counts()
```

```python
# Find the 7th movie of each genre
movies.groupby('Genre').nth(6)
```

```python
# Find movies of Horror Genre
movies.groupby('Genre').get_group('Horror')

# Alternatively -- movies[movies['Genre']=='Horror']
```

```python
movies.groupby('Genre').nunique()
```

```python
genres=movies.groupby('Genre')
```

```python
genres.agg(
    {
        'Runtime':['mean','min'],
        'IMDB_Rating':['mean','sum'],
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# Find highest rated movie in each genre
movies.groupby('Genre').apply(lambda x: x.loc[x['IMDB_Rating'].idxmax()][['Series_Title','IMDB_Rating']])
# The idxmax() function in pandas is used to return the index of the first occurrence of the maximum
# value in a Series or DataFrame
```

```python
genres['IMDB_Rating'].idxmax() # gives the index value of below series
```

```python
genres['IMDB_Rating'].max()
```

```python
# Find ranking of each movie in their own Genre according to IMDB rating
def rank(group):
    group['Rank']=group['IMDB_Rating'].rank(ascending=False)
    return group
genres.apply(rank)
```

```python
# find the best(in-terms of metascore(avg)) actor->genre combo
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(1)

# reset_index() converts a series into a dataframe
```
